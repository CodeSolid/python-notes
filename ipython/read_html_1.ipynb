# read_html demo
We import read_html and use it to read the tables from an FDIC web site into.  It returns a list of DataFrames (one per html table), so we get the first one out, then index by name into the 'City' column, which returns a panda Series object.  We get the length of the Series and the length of cities.unique(), which returns the unique cities.

Note:  both PyCharm and the web interface allow for previewing markdown using "Run Cell". 

```python
from pandas.io.html import read_html
```

```python
dfs = read_html('http://www.fdic.gov/bank/individual/failed/banklist.html')
```

```python
type(dfs)
```

```python
df1 = dfs[0]
```

```python
type(df1)
```

```python
cities = df1['City']
```

```python
type(cities)
```

```python
len(cities)
```

```python
len(cities.unique())
```

```python
cities[0:10]
```

```python
type(cities.unique())
```
